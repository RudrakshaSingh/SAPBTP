## Helper Functions (Task 5)

Defined early so they can be reused in later tasks.

````python
def calculate_result(marks):
    """Return Pass or Fail based on marks."""
    return "Pass" if marks >= 40 else "Fail"


def calculate_grade(marks):
    """Return grade based on marks range."""
    if marks >= 90:
        return "A+"
    if marks >= 75:
        return "A"
    if marks >= 60:
        return "B"
    if marks >= 40:
        return "C"
    return "Fail"


def calculate_average_marks(marks_list):
    """Return average of a list of marks."""
    if not marks_list:
        return 0.0
    return sum(marks_list) / len(marks_list)


def check_attendance_status(attendance):
    """Return attendance eligibility status."""
    return "Eligible" if attendance >= 75 else "Low Attendance"


def validate_marks_input(user_input):
    """Validate marks input from user. Returns valid marks or None."""
    try:
        marks = float(user_input)
        if marks < 0 or marks > 100:
            print("Invalid input. Please enter marks between 0 and 100.")
            return None
        return int(marks) if marks == int(marks) else marks
    except ValueError:
        print("Invalid input. Please enter marks between 0 and 100.")
        return None


def add_performance_columns(df):
    """Add Result, Grade, and Attendance_Status columns."""
    df = df.copy()
    df["Result"] = df["Marks"].apply(calculate_result)
    df["Grade"] = df["Marks"].apply(calculate_grade)
    df["Attendance_Status"] = df["Attendance_Percentage"].apply(check_attendance_status)
    return df
````

## Task 3: Loops

Using the marks list, print all marks and compute pass/fail counts, min, max, total, and average.

````python
marks = MARKS_LIST
passed_count = 0
failed_count = 0
total_marks = 0
highest = marks[0]
lowest = marks[0]

print("All marks:")
for mark in marks:
    print(mark)
    total_marks += mark
    if mark >= 40:
        passed_count += 1
    else:
        failed_count += 1
    if mark > highest:
        highest = mark
    if mark < lowest:
        lowest = mark

average_marks = total_marks / len(marks)

print(f"\nPassed Students: {passed_count}")
print(f"Failed Students: {failed_count}")
print(f"Highest Marks: {highest}")
print(f"Lowest Marks: {lowest}")
print(f"Total Marks: {total_marks}")
print(f"Average Marks: {average_marks}")
````

## Task 7: File Handling (Text File)

Create `student_summary.txt`, write the summary, then read and display it.

````python
total_students = len(MARKS_LIST)
average_marks = calculate_average_marks(MARKS_LIST)
highest_marks = max(MARKS_LIST)
lowest_marks = min(MARKS_LIST)
passed_students = sum(1 for m in MARKS_LIST if m >= 40)
failed_students = total_students - passed_students

summary_content = f"""Student Performance Summary
Total Students: {total_students}
Average Marks: {average_marks:.1f}
Highest Marks: {highest_marks}
Lowest Marks: {lowest_marks}
Passed Students: {passed_students}
Failed Students: {failed_students}
"""

with open(SUMMARY_TXT, "w", encoding="utf-8") as file:
    file.write(summary_content)

print(f"Summary written to: {SUMMARY_TXT.name}")
print("\nFile content:")
with open(SUMMARY_TXT, "r", encoding="utf-8") as file:
    print(file.read())
````

## Task 8: CSV File Handling

Write student records to `students.csv` and read them row by row.

````python
with open(STUDENTS_CSV, "w", newline="", encoding="utf-8") as file:
    writer = csv.DictWriter(file, fieldnames=CSV_COLUMNS)
    writer.writeheader()
    writer.writerows(STUDENT_DATA)

print(f"Student records written to: {STUDENTS_CSV.name}")
print("\nReading CSV file row by row:")
with open(STUDENTS_CSV, "r", encoding="utf-8") as file:
    reader = csv.DictReader(file)
    for row in reader:
        print(row)
````
